```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
videos = pd.read_csv("https://raw.githubusercontent.com/fedhere/FDSFE_FBianco/main/data/USyoutubes.csv")
```

```python
videos.head()
```

```python
videos_restrict = videos[["views", "likes", "dislikes", "comment_count"]]
```

```python
videos_restrict.tail()
```

# Find the video with the most likes in this dataset.

```python
# The maximum value of the likes column
videos_restrict["likes"].max()
```

```python
# The minimum value of the likes column
videos_restrict["likes"].min()
```

```python
# Storing the maximum value of the likes column in a variable.
max_likes = videos_restrict["likes"].max()

# Using that variable to index the dataset for the row which has the most number of likes.
videos[videos["likes"] == max_likes]
```

# Find the video with the most number of dislikes

```python
# Storing the maximum value of the dislikes column in a variable.
max_dislikes = videos_restrict["dislikes"].max()

# Using that variable to index the dataset for the row which has the most number of dislikes.
videos[videos["dislikes"] == max_dislikes]
```

# Find the video with the least number of likes

```python
# Storing the minimum value of the likes column in a variable.
min_likes = videos_restrict["likes"].min()

# Using that variable to index the dataset for the row which has the least number of likes.
videos[videos["likes"] == min_likes]
```
