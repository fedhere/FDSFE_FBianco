## conditionals and loops

##### if statement

```python
x = 2
if x == 1:
    print("1")
else:
    print(' not 1')
```

```python
x = 2

if x == 1:
    print("1")
elif x == 2:
    print("2")
elif x == 3:
    print("3")
else:
    print("others")
```

##### for loop

```python
for i in range(3):
    print(i)
```

```python
# use zip to loop over multiple variables

x = ['a', 'b', 'c', ]
y = [1, 2, 3]

for i, j in zip(x, y):
    print(i, j)
```

```python
# use enumerate get index of each loop

x = ['a', 'b', 'c', ]

for i, xi in enumerate(x):
        print(i, xi)
```

##### while loop

```python
x = 0
while x < 3:
    print(x)
    x += 1
```

## functions

```python
def add(a=0, b=0):
    """
    Add two numbers
    Args:
        a: number, default 0
        b: number, default 0

    Returns:
        return the sum

    """
    result = a + b
    return result
```

the bit between """ and """ is called a "docstring" and all functions should have one to describe input, output, and puspose

```python
add(1, 2)
```

## classes

Creating a new class creates a new type of object, like a complex variable that has inside of itself variables and methods

```python
class Car:
    def __init__(self,brand,**kwargs):
        self.brand = brand
        self.kwargs = kwargs

    def get_brand(self):
        return self.brand

    def get_prams(self):
        return self.kwargs
```

```python
# Instantiate
newCar = Car('BMW', price=10000, color='red', model='m')
```

after a class is defined, an object of the class is creted by "instantiating" it, defining all required values


```python
newCar.get_brand()
```

```python
newCar.get_prams()
```

##### Inheritance: a class can contain the initialization of another (parent) class

```python
class Coupe(Car):
    def __init__(self, brand,**kwargs):
        super().__init__(brand,**kwargs) # works in phthon 3
        #super(Coupe, self).__init__(brand,**kwargs)  # works in python 2/3

    def test(self):
        print(' class works')

    def __str__(self):
        # for print
        return self.brand
```

```python
# Instantiate
newCoupe = Coupe('BMW', price=10000, color='red', model='m')

newCoupe.test()
```

```python
newCoupe.__str__()
```
